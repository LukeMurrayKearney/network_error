```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
with open('../csv_output/error2_period1.json','r') as f:
    p1 = json.load(f)
with open('../csv_output/error2_period2.json','r') as f:
    p2 = json.load(f)
```

```python
ns = np.arange(20_000,101_000,1_000)
means1 = np.zeros(81)
for i,x in enumerate(p1[0]):
    means1[i] = x['mean']
means2 = np.zeros(81)
for i,x in enumerate(p2[0]):
    means2[i] = x['mean']
```

```python
plt.plot(ns, means1, label = 'Restricitons Easing')
plt.plot(ns,means2, label ='Schools open')
plt.ylabel('Reduction in Error')
plt.xlabel('Network size')
plt.legend()
plt.savefig('figs/reduction_in_error.png')
plt.show()
```
